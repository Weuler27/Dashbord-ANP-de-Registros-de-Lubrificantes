```python
#IMPORTA√á√ÉO DA BIBLIOTECA PANDAS
import pandas as pd 
```

```python
#IMPORTA√á√ÉO DOS DADOS PARA AN√ÅLISE
dados = pd.read_csv(r'C:\Users\user\OneDrive\Documentos\Projeto_Python\Projeto 1\dados-abertos-registro-produtos.csv' , sep =';' )
```

ANALISES A SEREM FEITAS:
- QUANTOS REGISTROS POR ANO FEITOS
- QUANTOS REGISTROS POR MARCA_COMERCIAL
- ANALISE CONTAGEM DOS DETENTORES
- ANALISE CONTAGEM DE TIPO DE EMPRESA
- ANALISE DE CONTAGEM APLICA√á√ÉO
- ANALISE DE CONTAGEM DE ORIGEM

```python
#LER O DATAFRAME AS 10 PRIMEIRAS LINHAS COMO PADR√ÉO
dados.head()
```

```python
#LER O DATAFRAME AS 10 ULTIMAS LINHAS
dados.tail(10)
```

```python
#VER O NOME DAS COLUNAS
dados.columns
```

```python
#QUANTOS REGISTROS POR ANO
registro_por_ano = dados['ANO'].value_counts().sort_index()
```

```python
#QUANTOS PRODUTOS POR MARCA (10 PRINCIPAIS)
produtos_por_marca = dados['MARCA_COMERCIAL'].value_counts().head(10)
```

```python
#QUANTOS DETENTORES
contagem_detentores = dados['DETENTOR'].value_counts().head(10)
```

```python
#QUANTOS TIPOS DE EMPRESA
tipos_empresa = dados['TIPO_EMPRESA'].value_counts() #quantidade de tipos de empresa.
dados_porcentagem_empresa = tipos_empresa / tipos_empresa.sum()*100 #tranforma√ß√£o em porcentagem
print(dados_porcentagem_empresa)
```

```python
#QUANTOS TIPOS DE APLIA√á√ÉO
tipos_aplicacao = dados['APLICACAO'].value_counts().head(15)
print(tipos_aplicacao)
```

```python
#QUANTOS TIPOS DE ORIGEM
tipos_origem = dados['ORIGEM'].value_counts() #quantidade de tipos de origem
dados_porcentagem_origem = tipos_origem / tipos_origem.sum()*100 #tranforma√ß√£o em porcentagem
print(dados_porcentagem_origem)
```

2_VISUALIZA√á√ÉO DOS DADOS

```python
#IMPORTA√á√ÉO DA BIBLIOTECA MATPLOTLIB 
import matplotlib.pyplot as plt
```

```python
#GR√ÅFICO DE LINNHA DE REGITROS POR ANO (USANDO MATPLOTIB)
plt.figure(figsize=(10,5))
plt.plot(registro_por_ano.index, registro_por_ano.values,
          marker = 'o', linestyle = '-', color = 'blue')
plt.title('Registros por Ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Registros', fontsize = 14)
plt.xticks(rotation = 45)
plt.grid(True)
plt.show()
```

```python
#GR√ÅFICO DE BARRAS HORIZONTAL DE QUANTIDADE DE PRODUTOS POR MARCA (MATPLOTLIB)
plt.figure(figsize=(10,5))
produtos_por_marca_2 = produtos_por_marca.sort_values(ascending=True) # colocar em ordem decrescente os valores
plt.barh(produtos_por_marca_2.index, produtos_por_marca_2.values, color = 'blue')
plt.title('Quantidade de Produtos por Marca', fontsize = 14)
plt.xlabel('Quantidade')
plt.ylabel('Marca')
plt.show()
```

```python
#GR√ÅFICO DE BARRAS HORIZONTAL DE QUANTIDADE DE DETENTORES (MATPLOTLIB)
plt.figure(figsize=(10,5))
contagem_detentores_2 = contagem_detentores.sort_values(ascending=True) #colocar em ordem decrescente os valores
plt.barh(contagem_detentores_2.index, 
         contagem_detentores_2.values, 
         color = 'blue')
plt.title('Quantidade de detentores', fontsize = 14)
plt.xlabel('Quantidade')
plt.ylabel('Detentores')
plt.show()
```

```python
#GR√ÅFICO DE PIZZA DE QUANTOS TIPOS DE EMPRESA (MATPLOLIB)
plt.figure(figsize=(10,5))
cmap = plt.get_cmap('winter')
cores = cmap([i / len(tipos_empresa) for i in range(len(tipos_empresa))])

plt.pie(dados_porcentagem_empresa,
        labels=dados_porcentagem_empresa.index,
        autopct='%1.1f%%', startangle=90, colors=cores, 
        wedgeprops={'edgecolor' : 'white' , 'linewidth' : 2}, 
        textprops={'color' : 'black', 'fontsize': 10})
plt.title('Distribui√ß√£o por Tipo de Empresa', fontsize = 14)
plt.show()
```

```python
#GR√ÅFICO DE BARRAS HORIZONTAL TIPOS DE APLICA√á√ÉO(MATPLOTLIB)
plt.figure(figsize=(10,5))
tipos_aplicacao_2 = tipos_aplicacao.sort_values(ascending=True)
plt.barh(tipos_aplicacao_2.index, 
         tipos_aplicacao_2.values,
         color = 'blue')
plt.title('Quantidade de Tipos de Aplica√ß√£o', fontsize = 14)
plt.xlabel('Quantidade')
plt.ylabel('Tipos de Aplica√ß√£o')
plt.show()
```

```python
#GR√ÅFICO DE PIZZA DE QUANTOS TIPOS DE ORIGEM (MATPLOLIB)
plt.figure(figsize=(10,5))
cmap = plt.get_cmap('winter')
cores = cmap([i / len(tipos_origem) for i in range(len(tipos_origem))])

plt.pie(dados_porcentagem_origem,
        labels=dados_porcentagem_origem.index,
        autopct='%1.1f%%', startangle=90, colors=cores, 
        wedgeprops={'edgecolor' : 'white' , 'linewidth' : 2}, 
        textprops={'color' : 'black', 'fontsize': 10})
plt.title('Distribui√ß√£o por Tipo de Origem', fontsize = 14)
plt.show()
```
